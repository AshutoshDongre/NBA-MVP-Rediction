```python
import pandas as pd
```

```python
mvps=mvps[["Player","Year","Pts Won","Pts Max","Share"]]
```

```python
mvps.head()
```

```python
players=pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players
```

```python
players["Player"]=players["Player"].str.replace("*","",regex=False)
```

```python
players.head()
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row=df[df["Tm"]=="TOT"]
        row["Tm"]=df.iloc[-1,:]["Tm"]
        return row

players=players.groupby(["Player","Year"]).apply(single_row)
```

```python
players.head()
```

```python
players.index=players.index.droplevel()
```

```python
players.index=players.index.droplevel()
```

```python
players.head()
```

```python
combined=players.merge(mvps,how="outer",on=["Player","Year"])
```

```python
combined
```

```python
combined[["Pts Won","Pts Max","Share"]]=combined[["Pts Won","Pts Max","Share"]].fillna(0)
```

```python
combined
```

```python
teams=pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams=teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"]=teams["Team"].str.replace("*","",regex=False)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames={}
with open("nicknames.csv") as f:
    lines=f.readlines()
    for line in lines[1:]:
        abbrev,name=line.replace("\n","").split(",")
        nicknames[abbrev]=name
```

```python
nicknames
```

```python
combined["Team"]=combined["Tm"].map(nicknames)
```

```python
combined.loc[combined["Team"].isna()]
```

```python
combined
```

```python
stats=combined.merge(teams,how="inner",on=["Team","Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

```python
stats=stats.apply(pd.to_numeric,errors="ignore")
```

```python
stats["GB"]=stats["GB"].str.replace("—","",regex=False)
```

```python
stats["GB"].unique()
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
stats.corr()["Share"]
```
